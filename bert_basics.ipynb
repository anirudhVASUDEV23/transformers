```python
from transformers import BertTokenizer,BertModel
```

```python
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
tokenizer("I love Apple", return_tensors='pt')##101 is cls and 102 is sep
```

```python
tokens=tokenizer(["I love Apple", "I love Orange and Banana"], return_tensors='pt', padding=True)
tokens
```

```python
tokens=tokenizer(["I love Apple", "I love Orange and Banana"], return_tensors='pt', padding="max_length", max_length=10)
tokens
```

```python
tokens=tokenizer(["I love Apple", "you are what you eat"], return_tensors='pt', padding="max_length", max_length=5, truncation=True)
tokens
```

```python
tokens=tokenizer("I love Apple", return_tensors='pt')
tokens
```

```python
model=BertModel.from_pretrained('bert-base-uncased')
outputs=model(**tokens)##outputs the contextualized embeddings
```

```python
outputs[0][0].shape
```

```python
outputs['last_hidden_state']
```

```python
outputs['last_hidden_state'].shape##total of 5 vectors,each is a contextualized embedding for tokens in the sentence "i love apple" along with cls and sep tokens and each embedding has 768 dimensions
```

```python
outputs['last_hidden_state'][0,0,:]##cls token embedding
```

```python
outputs['last_hidden_state'][0,1,:]##i token contextualized embedding
```

```python
outputs['pooler_output']##is single vector representation of the sentence "I love Apple" obtained by applying a linear transformation and tanh activation on the cls token embedding
outputs['pooler_output'].shape##single vector of 768 dimensions
```

```python
outputs['pooler_output'][0,:]##single vector representation of the sentence "I love Apple"
```

```python
##CLS IS ALSO EMBEDDING FOR THE WHOLE SENTENCE
###POOLER OUTPUT IS ALSO EMBEDDING FOR THE WHOLE SENTENCE,WHICH IS OBTAINED BY PASSING CLS TOKEN THROUGH A NEURAL NETWORK AND TAKING TANH OF IT which is activation function
```
